# **Model**

```python
!pip install pyomo
import sys
import numpy as np
from pyomo import environ as pe
from pyomo.environ import *
```

```python
# สร้างโมเดล
model = ConcreteModel()
# Sets
model.C = Set(initialize=C.keys())  # รายวิชา
model.R = Set(initialize=R.keys())  # ห้องเรียน
model.T = Set(initialize=T.keys())  # ช่วงเวลา
model.P = Set(initialize=P.keys())  # ผู้สอน
model.S = Set(initialize=S.keys())  # นักศึกษา
model.T_M = Set(initialize=range(1, 19))  # วันจันทร์
model.T_T = Set(initialize=range(19, 37))  # วันอังคาร
model.T_W = Set(initialize=range(37, 55))  # วันพุธ
model.T_Th = Set(initialize=range(55, 73))  # วันพฤหัสบดี
model.T_F = Set(initialize=range(73, 91))  # วันศุกร์
model.c_p = Set(dimen=2)
# Variables
model.x_crt = Var(model.C, model.R, model.T, domain=Binary)
```

```python
model.obj = Objective(expr=summation(model.x_crt), sense=maximize)
```

# **solve**

```python
!sudo apt-get install glpk-utils
solver = pe.SolverFactory('glpk', executable='/usr/bin/glpsol')
solution = solver.solve(model)
```

```python
from pyomo.opt import SolverFactory

# กำหนด Solver

opt = pe.SolverFactory('glpk')
opt.solve(model, tee=True)

# แสดงผลลัพธ์
model.display()
```

```python
# บันทึกโมเดลเป็นไฟล์ .lp
model.write('model.lp', io_options={'symbolic_solver_labels': True})

from google.colab import files
files.download('model.lp')
```
