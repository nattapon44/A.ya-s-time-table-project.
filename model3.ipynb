```python
import numpy as np

def calculate_teacher_demand(c):
    demands = {}

    for key in range(1, 6):
        course_name = c[key]["courseName"]
        teacher_count = len(c[key]["teachers"])
        hours_per_week = c[key]["hoursPerWeek"]
        course_type = c[key]["type"]

       # print(f"\nCourse: {course_name}")
       # print(f"Number of Teachers: {teacher_count}")

        if teacher_count == 1:
            teacher_id = list(c[key]["teachers"].keys())[0]
            teacher_matrix = get_teacher_weights(teacher_id, P)
            demands[course_name] = teacher_matrix
            print(f"Teacher {teacher_id}'s Weight Matrix:")
            print(teacher_matrix)
        elif teacher_count == 2:
            # Find the element-wise minimum of the two weight matrices
            teacher_ids = list(c[key]["teachers"].keys())
            weights1 = get_teacher_weights(teacher_ids[0], P)
            weights2 = get_teacher_weights(teacher_ids[1], P)

            min_weights = np.minimum(weights1, weights2)

            demands[course_name] = min_weights
            print(f"Minimum Weight Matrix:")
            print(min_weights)
        else:
            print("Unsupported number of teachers (more than 2).")

    return demands

# ใช้ฟังก์ชันเพื่อคำนวณความต้องการสอนของผู้สอน
u_crt = calculate_teacher_demand(C)
```

```python
!pip install pyomo
import sys
import numpy as np
from pyomo import environ as pe
from pyomo.environ import *
```

```python
# สร้างโมเดล
model = ConcreteModel()
# Sets
model.C = Set(initialize=C.keys())  # รายวิชา
model.R = Set(initialize=R.keys())  # ห้องเรียน
model.T = Set(initialize=T.keys())  # ช่วงเวลา
model.P = Set(initialize=P.keys())  # ผู้สอน
model.S = Set(initialize=S.keys())  # นักศึกษา
model.D = Set(initialize=TD.keys())
model.T_M = Set(initialize=range(1, 19))  # วันจันทร์
model.T_T = Set(initialize=range(19, 37))  # วันอังคาร
model.T_W = Set(initialize=range(37, 55))  # วันพุธ
model.T_Th = Set(initialize=range(55, 73))  # วันพฤหัสบดี
model.T_F = Set(initialize=range(73, 91))  # วันศุกร์
model.C_p = Set(initialize=P.keys())
model.C_s = Set(initialize=S.keys())
model.R_c = Set(initialize=C.keys())
model.T_p = Set(initialize=P.keys())
model.T_tc = Set(initialize=T.keys())
```

```python
model.u_crt = Param(model.P, initialize=teacher_demands, mutable=True, within=NonNegativeReals)
```

```python
model.u_crt = Param(model.C, model.R, model.T, initialize=u_crt)
```

```python
# Parameters
#model.u_crt = Param(model.C, model.R, model.T,within=NonNegativeReals)  # น้ำหนักความต้องการสอน
model.cap_r = Param(model.R, within=PositiveIntegers)  # ความจุของห้องเรียน
model.cap_c = Param(model.C, within=PositiveIntegers)  # ความจุของรายวิชา
model.h_c = Param(model.C, initialize=h_c ,within=PositiveIntegers)  # จำนวนคาบการสอนทั้งหมดในหนึ่งสัปดาห์ของรายวิชา c
model.k_p = Param(model.P, within=PositiveIntegers)  # จำนวนชั่วโมงที่สามารถสอนได้ต่อวัน
model.k_s = Param(model.S, within=PositiveIntegers)  # จำนวนชั่วโมงที่สามารถเรียนได้ต่อวัน
model.k_c = Param(model.C, within=PositiveIntegers)  # จำนวนครั้งที่เรียนในรายวิชา c ในหนึ่งสัปดาห์
model.h_cd = Param(model.C, model.D, within=PositiveIntegers)  # ช่วงเวลาของวิชา c ที่เรียนในแต่ละวัน d
```

```python
model.x_crt = Var(model.C, model.R, model.T, within=Binary)
```

```python
# Variables
model.x_crt = Var(model.C, model.R, model.T, within=Binary)  # ตัวแปร binary ที่ระบุว่ารายวิชา c ได้รับการมอบหมายในห้อง r ในช่วงเวลา t
model.z_scrt = Var(model.S, model.C, model.R, model.T, within=Binary)  # ตัวแปร binary ที่ระบุว่านักศึกษา s เรียนในรายวิชา c ถูกมอบหมายให้เรียนห้อง r ในช่วงเวลา t
model.w_c = Var(model.C, within=Binary)  # ตัวแปร binary ที่ระบุว่ารายวิชา c ได้รับการจัดตาราง
model.w_c_d = Var(model.C, model.D, within=Binary)  # ตัวแปร binary ที่ระบุว่ารายวิชา c ถูกจัดตารางในวัน d
model.v_cd = Var(model.C, model.D, within=Binary)  # ตัวแปร binary ที่ระบุว่ารายวิชา c ที่เรียนในวัน d ∈ D
```

```python
model.objective = Objective(expr=sum(model.x_crt[c, r, t] for c in model.C for r in model.R_c for t in model.T), sense=maximize)
```

```python
# Objective Function
model.obj = Objective(
    expr=sum(model.u_crt[c, r, t] * model.x_crt[c, r, t] for c in model.C for r in model.C for t in model.T),
    sense=maximize
)
```

```python
# Objective Function
model.obj = Objective(
    expr=sum((model.u_crt[c, r, t] / model.h_c[c]) * model.x_crt[c, r, t] for c in model.C for r in model.C for t in model.T),
    sense=maximize
)
```

```python
# Constraints
# ข้อจำกัดที่ 1
def con1(model, t, r):
    return sum(model.x_crt[c, r, t] for c in model.C) <= 1
model.con1 = Constraint(model.T, model.R, rule=con1)
```

```python
# ข้อจำกัดที่ 2
def con2(model, p, t):
    return sum(model.x_crt[c, r, t] for c in model.C_p for r in model.R_c[c] if p in model.T_p[c]) <= 1
model.con2 = Constraint(model.P, model.T, rule=con2)
```

```python
# 3. จำนวนนักศึกษาต้องไม่มากกว่าความจุห้องเรียน
def con3(model, r, t):
    return sum(model.z_scrt[s, c, r, t] for s in model.S for c in model.C for r in model.R for t in model.T) <= model.Cap_r[r]
model.con3 = Constraint(model.R, model.T, rule=con3)
```

```python
# 4. จำนวนนักศึกษาต้องไม่มากกว่าความจุรายวิชา
def con4(model, c, t):
    return sum(model.z_scrt[s, c, r, t] for s in model.S for c in model.C for r in model.R for t in model.T) <= model.Cap_c[c]
model.con4 = Constraint(model.C, model.T, rule=con4)
```
